```python
import pandas as pd
import plotly.graph_objs as go
import plotly.offline as pyo
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/Pierian-Data/Plotly-Dashboards-with-Dash/master/SourceData/nst-est2017-alldata.csv')
df.head()
```

```python
df2=df[df['DIVISION']=='1']
```

```python
df2.set_index('NAME',inplace=True)
```

```python
df2
```

```python
list_pop_col=[col for col in df2.columns if col.startswith('POP')]
```

```python
df2=df2[list_pop_col]
```

```python
df2
```

```python
data=[go.Scatter(x=df2.columns,y=df2.loc[name],mode='lines',name=name) for name in df2.index]
```

```python
pyo.plot(data)
```
